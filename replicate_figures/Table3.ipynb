# Calculate TCRs

```python
# Calculate from scratch
TCR_DICT = calculate_tcr_time_series.create_tcr_datasets(tcr_types=['RAD','TOT','PHYS','TOT-RAD'],
                                                         average_types=['global','land'],
                                                         save_tcr_dict=True,
                                                         recalculate_TCRs=True)
```

```python
# Load precalculated
DIR_TCR_DICT = directory_information.DIR_TCR_DICT
TCR_DICT = pickle.load(open(DIR_TCR_DICT+'TCR_dict.pickle', "rb"))
```

### Calculate TCRs

```python
TCR_global = make_tcr_dataset.make_tcr_dataset(TCR_DICT,
                                              end_yr=70,
                                              varname='tas',
                                              average_type='global',
                                              tcr_types=['RAD','TOT','TOT-RAD'])
TCR_land = make_tcr_dataset.make_tcr_dataset(TCR_DICT,
                                              end_yr=70,
                                              varname='tas',
                                              average_type='land',
                                              tcr_types=['RAD','TOT','TOT-RAD'])
TCR_ocean = make_tcr_dataset.make_tcr_dataset(TCR_DICT,
                                              end_yr=70,
                                              varname='tas',
                                              average_type='ocean',
                                              tcr_types=['RAD','TOT','TOT-RAD'])
T140_global = make_tcr_dataset.make_tcr_dataset(TCR_DICT,
                                              end_yr=130,
                                              varname='tas',
                                              average_type='global',
                                              tcr_types=['RAD','TOT','TOT-RAD'])
T140_land = make_tcr_dataset.make_tcr_dataset(TCR_DICT,
                                              end_yr=130,
                                              varname='tas',
                                              average_type='land',
                                              tcr_types=['RAD','TOT','TOT-RAD'])
T140_ocean = make_tcr_dataset.make_tcr_dataset(TCR_DICT,
                                              end_yr=130,
                                              varname='tas',
                                              average_type='ocean',
                                              tcr_types=['RAD','TOT','TOT-RAD'])
```

# Construct Table

```python
cmip_cutoff=8
TCR_global_CMIP5 = TCR_global[0:cmip_cutoff]
TCR_global_CMIP6 = TCR_global[cmip_cutoff:len(TCR_global)]
TCR_land_CMIP5 = TCR_land[0:cmip_cutoff]
TCR_land_CMIP6 = TCR_land[cmip_cutoff:len(TCR_global)]
T140_global_CMIP5 = T140_global[0:cmip_cutoff]
T140_global_CMIP6 = T140_global[cmip_cutoff:len(TCR_global)]
T140_land_CMIP5 = T140_land[0:cmip_cutoff]
T140_land_CMIP6 = T140_land[cmip_cutoff:len(TCR_global)]
```

```python
TCR_ocean_CMIP5 = TCR_ocean[0:cmip_cutoff]
TCR_ocean_CMIP6 = TCR_ocean[cmip_cutoff:len(TCR_global)]
T140_ocean_CMIP5 = T140_ocean[0:cmip_cutoff]
T140_ocean_CMIP6 = T140_ocean[cmip_cutoff:len(TCR_global)]
```

```python
TCR_FULL_PI = [TCR_global_CMIP6['TOT'].std(),
               TCR_land_CMIP6['TOT'].std(),
               TCR_ocean_CMIP6['TOT'].std(),
               TCR_global_CMIP5['TOT'].std(),
               TCR_land_CMIP5['TOT'].std(),
               TCR_ocean_CMIP5['TOT'].std()]
TCR_RAD_PI = [TCR_global_CMIP6['RAD'].std(),
               TCR_land_CMIP6['RAD'].std(),
               TCR_ocean_CMIP6['RAD'].std(),
               TCR_global_CMIP5['RAD'].std(),
               TCR_land_CMIP5['RAD'].std(),
               TCR_ocean_CMIP5['RAD'].std()]
T140_FULL_PI = [T140_global_CMIP6['TOT'].std(),
               T140_land_CMIP6['TOT'].std(),
               T140_ocean_CMIP6['TOT'].std(),
               T140_global_CMIP5['TOT'].std(),
               T140_land_CMIP5['TOT'].std(),
               T140_ocean_CMIP5['TOT'].std()]
T140_RAD_PI = [T140_global_CMIP6['RAD'].std(),
               T140_land_CMIP6['RAD'].std(),
               T140_ocean_CMIP6['RAD'].std(),
               T140_global_CMIP5['RAD'].std(),
               T140_land_CMIP5['RAD'].std(),
               T140_ocean_CMIP5['RAD'].std()]
```

```python
df = pd.DataFrame(['CMIP6','CMIP6','CMIP6','CMIP5','CMIP5','CMIP5'],
                  columns=['CMIP_Phase'])
df['Region'] = ['Global','Land','Ocean','Global','Land','Ocean']
df['TCR_FULL-PI'] = TCR_FULL_PI
df['TCR_RAD-PI'] = TCR_RAD_PI
df['TCR_RADpct'] = df['TCR_RAD-PI']*100/df['TCR_FULL-PI']
df['T140_FULL-PI'] = T140_FULL_PI
df['T140_RAD-PI'] = T140_RAD_PI
df['T140_RADpct'] = df['T140_RAD-PI']*100/df['T140_FULL-PI']
```
