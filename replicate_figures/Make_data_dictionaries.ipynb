### Check what is in dictionary

```python
import pandas as pd
from plants_and_TCR.analysis_parameters import get_CMIP_info
from plants_and_TCR.make_data_dictionary import save_and_open_dict
from plants_and_TCR.analysis_parameters import directory_information
from plants_and_TCR.analyze_data import grab_cmip_dataset
PATH_DICT = directory_information.DIR_DATA_DICTIONARIES
```

```python
cmip_dict = save_and_open_dict.open_dict(path=PATH_DICT, dict_name='cmip_dict')
modelnames = get_CMIP_info.get_modelnames_short('CMIP5and6')
runnames =['1pctCO2-rad','1pctCO2-bgc','1pctCO2','piControl']
data_availability = pd.DataFrame(index=modelnames,columns=runnames)
```

```python
VARNAME='tas'
for runname in runnames:
    for modelname in modelnames:
        ds = grab_cmip_dataset.grab_cmip_dataset(cmip_dict, modelname, runname, VARNAME)
        if ds is None:
            data_availability[runname][modelname]=False
        else:
            data_availability[runname][modelname]=True
```

```python
def highlight_true(s):
    is_true = (s == True)
    return ['background-color: lemonchiffon' if v else '' for v in is_true]
```

```python
data_availability.style.apply(highlight_true)
```
